## Global variable
    Variables that are created inside of a function but can be used out of it too are known as global variables.

```python
def function():
    global variable
    variable = 10

function()
print(variable)

#if we uncomment the global b statement  the cell will result to an NameError saying 'variable' is not defined
```

## Multi threading
    Multithreading is defined as the ability of a processor to execute multiple threads concurrently.(reducing time complexity)

```python
import time
import threading

arr =[x for x in range(100)]
def append_nums(l):
    print("calculation of squares")
    for i in l:
        print(f"The square of {i} is : ",i**2)
    
def append_nums2(l):
    print("calculation of cubes")
    for i in l:
        print(f"The cube of {i} is : ",i**3)
        

t = time.time()
t1 = threading.Thread(target=append_nums, args=(arr, )) #creating threads to run simultaneously
t2 = threading.Thread(target=append_nums2, args=(arr, ))#defining the work of threads

t1.start() #starting the thread
t2.start() #starting the thread

#t1.join()#when you need to perform dependent tasks synchronously(HERE WE ARE PERFORMING TWO INDEPENDENT TASKS )
#t2.join()#when you need to perform dependent tasks synchronously

print("Done in :",(time.time()-t))#calc the time in seconds
```
